### **Q7) Show all courses of Graphic Design where the price is below 100.**

```python
data.head(1)
```

```python
data.dtypes
```

```python
data.loc[data["subject"] == "Graphic Design", "price_num"] = data["price"].replace("Free", 0).astype(int)
filtered_dataset = data[data["price_num"] < 100]
filtered_dataset
```

### **Q8) List out all the courses that are related with "Python".**

```python
data[data["course_title"].str.contains("Python")]
```

### **Q9) What are courser published in year 2015".**

```python
data.head(2)
```

```python
data['timestamp'] = pd.to_datetime(data["published_timestamp"])
data["year"] = data["timestamp"].dt.year
data[data["year"] == 2015]
```

### **Q10) What are the Max number of Subscribers for Each Level of courses".**

```python
data.head(3)
```

```python
data.groupby("level").num_subscribers.max().sort_values(ascending=False)
```
