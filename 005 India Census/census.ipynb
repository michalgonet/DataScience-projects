### **Q4) Calcualte state-wise:**
1). Total number of population

```python
data.groupby("State_name").Population.sum().sort_values(ascending=False)
```

2). Total number of population with different religions

```python
data.columns
```

```python
religions = ['Hindus', 'Muslims', 'Christians', 'Sikhs', 'Buddhists', 'Jains']
data.groupby('State_name')[religions].sum().sort_values(by= 'Hindus')
```
