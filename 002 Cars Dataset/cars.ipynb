# Cars Dataset
In this project data of different cars is given with their specifications. The data is in CVS format, and it is analyzed using Pandas DataFrame

```python
import pandas as pd
pd.set_option('display.expand_frame_repr', False) # for showing width of the colums better
```

```python
cars = pd.read_csv('cars_dataset.csv')
```

### **Q1) Find all Null Value in the dataset. If there is any null value in any column, then fill it with the mean of that column.**

```python
# how many nulls are in each column
cars.isnull().sum()
# Check if there are records where all values are NaN
cars[cars.isnull().any(axis=1)]
# 4 rows have only NaN values this rows would be removed from dataframe
cars = cars.dropna(how='all')
cars[cars.isnull().any(axis=1)]
# now only two rows cointain NaN values for Cylinder column
cars["Cylinders"].fillna(cars["Cylinders"].mean(), inplace=True)

# Check for NaN values again 
cars.isnull().sum()
```

### **Q2) Check what are the different types of Make are there in dataset and what is the count (occurence) of each Make in the data.**

### **Q3) Show all records where Origin is Asia or Europe.**

```python
cars.columns
sol1 = cars[(cars["Origin"]=="Asia") | (cars["Origin"]=="Europe")]
sol2 = cars[cars["Origin"].isin(["Asia","Europe"])]

print(f'There are {len(sol1)} cars using solution 1 and {len(sol2)} using solution 2')
```
